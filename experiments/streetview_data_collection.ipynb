### OECD definitions (copied)

Urban population by city size is determined by population density and commuting patterns; this better reflects the economic function of cities in addition to their administrative boundaries.

Urban areas in OECD countries are classified as:

* large metropolitan areas if they have a population of 1.5 million or more;
* metropolitan areas if their population is between 500 000 and 1.5 million;
* medium-size urban areas if their population is between 200 000 and 500 000; and,
* small urban areas if their population is between 50 000 and 200 000.
This indicator is measured as a percentage of the national population.

```python
def categorize_city_oecd(population):
    if population >= 1_500_000:
        return 'large_metropolitan'
    elif 500_000 <= population < 1_500_000:
        return 'metropolitan'
    elif 200_000 <= population < 500_000:
        return 'medium_urban'
    else:
        return 'small_urban'
```

```python
def sample_cities(df, sample_size=200):
    continents = ['Asia', 'North America', 'Africa', 'Europe', 'South America', 'Oceania']

    cities_per_continent = sample_size//len(continents)
    print(cities_per_continent)

    sampled_cities = []

    for continent in continents:
        continent_df = df[df['Continent'] == continent].copy()

        if len(continent_df) == 0:
            continue

        print(len(continent_df))


        continent_df['population'] = pd.to_numeric(continent_df['population'], errors='coerce')
        continent_df['population_group'] = continent_df['population'].apply(categorize_city_oecd)

        # Excluding 'small urban' classification for now
        for group in ['large_metropolitan', 'metropolitan', 'medium_urban']:
            group_df = continent_df[continent_df['population_group'] == group]
            group_sample_size = cities_per_continent//3

            sampled_cities.append(group_df.sample(min(group_sample_size, len(group_df))))


    result_df = pd.concat(sampled_cities, ignore_index=True)
    result_df = result_df.sample(frac=1).reset_index(drop=True)

    return result_df
```

```python
def random_point_in_radius(lat, lng, radius_km):
    R = 6371
    # convert km to radians
    radius_deg = radius_km / R

    # generate random polar coords
    u, v = random.random(), random.random()
    w = radius_deg * math.sqrt(u)
    t = 2 * math.pi * v

    # conver to cartesian coords
    x = w * math.cos(t)
    y = w * math.sin(t)

    new_x = x / math.cos(math.radians(lat))
    new_lat = lat + y
    new_lng = lng + new_x
    return new_lat, new_lng
```

```python
def download_streetview_image(lat, lng, city_name, index, fov=90, max_attempts=5, size=640):
    radius_km = 10  # look for random point within 10km radius

    for attempt in range(max_attempts):
        new_lat, new_lng = random_point_in_radius(lat, lng, radius_km)
        random_heading = random.randint(0, 359)

        # change heading and pitch to move viewpoint
        url = f"https://maps.googleapis.com/maps/api/streetview?size={size}x{size}&location={new_lat},{new_lng}&fov={fov}&heading={random_heading}&pitch=0&key={get_api_key()}"
        url += "&return_error_code=true" # returns error message instead of blank image
        url += "&source=outdoor"

        response = requests.get(url)

        if response.status_code == 200:
            img = Image.open(BytesIO(response.content))

            # crop google watermark
            width, height = img.size
            img = img.crop((0, 0, width, height - 30))
            city_name = city_name.replace(' ', '_')
            filename = f"imgs_v2/{index}.jpg"
            img.save(filename)
            print(f"Success! Downloaded image for {city_name} at {new_lat}, {new_lng}")
            return {"filename": filename, "city_name": city_name, "lat": new_lat, "lng": new_lng, "fov": random_heading}
        else:
            print(f"Attempt {attempt}. Error downloading image for {city_name} at {new_lat}, {new_lng}")

    print(f"Failed to download image for {city_name}.")
    return None
```

```python
coordinates_df = pd.read_csv('img_coordinates_v2_updated.csv')

img_coordinates = []
counter = 0
for index, row in coordinates_df.iterrows():
    result = download_streetview_image(row['lat'], row['lng'], row['city_name'], counter, max_attempts=10)
    if result:
        img_coordinates.append(result)
        counter += 1

print(len(img_coordinates))
img_coordinates = pd.DataFrame(img_coordinates)
img_coordinates.to_csv("img_coordinates_v2.csv", index=False)
```

```python
import pyexiv2

def get_all_data(image_path):
    # Open the image
    image = pyexiv2.Image(image_path)
    exif_data = image.read_exif()
    iptc_data = image.read_iptc()
    xmp_data = image.read_xmp()
    image.close()

    return {
        'EXIF': exif_data,
        'IPTC': iptc_data,
        'XMP': xmp_data
    }


metadata = get_all_data('imgs_v2/0_Huacho.jpg')
print(metadata)
```
